```python
import numpy as np
from stl import mesh
import k3d

filename = "test.stl"
e2c =[None]*12
e2c[0] = [0, -1, -1]
e2c[1] = [-1, -1, 0]
e2c[2] = [-1, 0, -1]
e2c[3] = [0, 1, 1]
e2c[4] = [1, 1, 0]
e2c[5] = [1, 0, 1]
e2c[6] = [1, 0, -1]
e2c[7] = [1, -1, 0]
e2c[8] = [0, -1, 1]
e2c[9] = [-1, 0, 1]
e2c[10] = [-1, 1, 0]
e2c[11] = [0, 1, -1]

def poly2tr(a):
    return [[a[0],a[i+1],a[i+2]] for i in range(len(a)-2)]
```

```python
tr = tlt[39]
verticesArray = np.array([[e2c[f] for f in e] for p in tr for e in poly2tr(p)])
solid = mesh.Mesh(np.zeros(verticesArray.shape[0], dtype=mesh.Mesh.dtype))
solid.vectors[:] = verticesArray
solid.save(filename)

with open(filename, 'rb') as stl:
    data = stl.read()

plt_test = k3d.stl(data, color=0xe3dac9)

plot = k3d.plot()
plot += plt_test
plot.display()
```
